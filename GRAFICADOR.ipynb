```python
import matplotlib.pyplot as pyplot
import numpy as np
import pandas as pd
import os
```

##Directorios y carga de archivos

```python
import os

# Entrada de datos (Mediciones)
DATA_DIR = "DATA"
# Guarda de Mediciones
MEASURE_DIR = "mic_behringher_ecm8000"
#Salida de graficos
OUTPUT_DIR = "OUTPUT"
output_path = os.path.join(OUTPUT_DIR, MEASURE_DIR)
os.makedirs(output_path, exist_ok=True)

# Lista de archivos txt
files = os.listdir(os.path.join(DATA_DIR, MEASURE_DIR))
print(files[0])
```

ANALISIS DE DATOS


```python
# Grilla Dataframe
path_files = os.path.join(DATA_DIR, MEASURE_DIR, files[0])
df = pd.read_csv(path_files, delimiter = "\t", header=1)
df
```

GRAFICO DE RESULTADOS

```python
import matplotlib.pyplot as plt

#Limite de ejes
y_ticks = np.arange(-12, 13, 3)
x_ticks = [20,100,1000,10000,20000]

# GrÃ¡fico del Dataframe, Frecuencia vs Amplitud
plt.figure(figsize=(10,6))
plt.plot(df['Frequency (Hz)'], df['Magnitude (dB)'])
plt.title('Respuesta en Frecuencia del micrÃ³fono Behringer ECM8000 a 0Â°', fontsize=16)

# Etiquetas de los ejes y fuentes
plt.xscale('log')
plt.xlim(20, 20000)
plt.ylim(-12, 12)
plt.yticks(y_ticks)
plt.xticks(x_ticks, x_ticks)
plt.xlabel('Frecuencia (Hz)', fontsize=14, fontname='Arial')
plt.ylabel('Amplitud (dB)', fontsize=14, fontname='Arial')
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.savefig(os.path.join(output_path, 'behringer_ecm8000_0.png'), dpi=300)
plt.show()
```
